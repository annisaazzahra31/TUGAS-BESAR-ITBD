```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **1. Tren Transaksi per Tahun**

```python
df_tren = pd.read_csv("tren_tahun.csv")

plt.figure(figsize=(6,4))
plt.plot(df_tren["Year"], df_tren["count"], marker="o")
plt.title("Tren Jumlah Transaksi per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Transaksi")
plt.grid(True)
plt.show()
```

Visualisasi ini menunjukkan perkembangan jumlah transaksi dari tahun ke tahun. Terlihat adanya peningkatan signifikan dari tahun 2023 ke 2024, yang mengindikasikan pertumbuhan aktivitas transaksi pelanggan. Namun, pada tahun 2025 terjadi penurunan jumlah transaksi yang cukup tajam, kemungkinan disebabkan oleh data yang belum mencakup satu tahun penuh.

# **2. Tren Jumlah Transaksi per Bulan**

```python
df_bulan = pd.read_csv("tren_bulan.csv")

plt.figure(figsize=(7,4))
plt.plot(df_bulan["Month"], df_bulan["count"], marker="o")
plt.title("Tren Jumlah Transaksi per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Transaksi")
plt.grid(True)
plt.show()
```

Grafik ini menampilkan pola transaksi bulanan. Jumlah transaksi cenderung stabil di setiap bulan dengan fluktuasi ringan. Beberapa bulan menunjukkan peningkatan transaksi, yang dapat mengindikasikan adanya faktor musiman atau periode promosi tertentu yang memengaruhi perilaku pembelian pelanggan.

# **3. Jumlah Transaksi per Kategori Produk**

```python
df_kat = pd.read_csv("kategori.csv")

plt.figure(figsize=(8,4))
bars = plt.barh(df_kat["Product_Category"], df_kat["count"])

plt.title("Jumlah Transaksi per Kategori Produk")
plt.xlabel("Jumlah Transaksi")
plt.ylabel("Kategori Produk")
plt.gca().invert_yaxis()

# kasih angka di ujung bar
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 50,
        bar.get_y() + bar.get_height()/2,
        f"{int(width)}",
        va="center"
    )

plt.show()
```

Visualisasi ini menunjukkan distribusi jumlah transaksi berdasarkan kategori produk. Kategori Toys, Electronics, dan Sports memiliki jumlah transaksi tertinggi, yang menunjukkan bahwa produk-produk tersebut paling diminati oleh pelanggan. Sementara itu, kategori Beauty memiliki jumlah transaksi paling rendah dibandingkan kategori lainnya.

# **4. Total Revenue per Kategori Produk**

```python
df_rev_kat = pd.read_csv("kategori_revenue.csv")

plt.figure(figsize=(8,4))
bars = plt.barh(
    df_rev_kat["Product_Category"],
    df_rev_kat["Total_Revenue"]
)

plt.title("Total Revenue per Kategori Produk")
plt.xlabel("Total Revenue")
plt.ylabel("Kategori Produk")
plt.gca().invert_yaxis()

#label revenue dengan format jutaan
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 50000,
        bar.get_y() + bar.get_height()/2,
        f"{width/1_000_000:.2f} Jt",
        va="center"
    )

plt.show()
```

Grafik di atas menunjukkan total pendapatan (revenue) untuk setiap kategori produk. Meskipun perbedaan revenue antar kategori tidak terlalu signifikan, kategori Sports dan Toys menghasilkan pendapatan tertinggi dibandingkan kategori lainnya. Hal ini menunjukkan bahwa kontribusi pendapatan relatif merata di seluruh kategori produk, tanpa adanya dominasi yang sangat mencolok.

# **5. Jumlah Transaksi per Negara**

```python
df_country = pd.read_csv("country.csv")

# ambil top 5 aja
df_top_country = df_country.head(5)

plt.figure(figsize=(8,4))
bars = plt.barh(df_top_country["Country"], df_top_country["count"])

plt.title("Top 5 Negara dengan Jumlah Transaksi Tertinggi")
plt.xlabel("Jumlah Transaksi")
plt.ylabel("Negara")
plt.gca().invert_yaxis()

for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 30,
        bar.get_y() + bar.get_height()/2,
        f"{int(width)}",
        va="center"
    )

plt.show()

```

Visualisasi ini difokuskan pada lima negara dengan jumlah transaksi tertinggi untuk memperjelas perbedaan antar negara. Terlihat bahwa Canada dan Mexico memiliki jumlah transaksi sedikit lebih tinggi dibandingkan negara lain, meskipun secara keseluruhan distribusi transaksi antar negara relatif merata.

# **6. Total Revenue per Negara**

```python
df_rev_country = pd.read_csv("country_revenue.csv")

# urutkan dari revenue terbesar
df_rev_country = df_rev_country.sort_values(
    "Total_Revenue", ascending=False
)

plt.figure(figsize=(8,5))
bars = plt.barh(
    df_rev_country["Country"],
    df_rev_country["Total_Revenue"]
)

plt.title("Total Revenue per Negara")
plt.xlabel("Total Revenue")
plt.ylabel("Negara")
plt.gca().invert_yaxis()

# kasih label angka (jutaan)
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 50000,
        bar.get_y() + bar.get_height()/2,
        f"{width/1_000_000:.2f} Jt",
        va="center"
    )

plt.show()
```

Grafik di atas menunjukkan distribusi total pendapatan berdasarkan negara pelanggan. Terlihat bahwa kontribusi revenue dari setiap negara relatif merata, dengan Prancis, Kanada, dan Amerika Serikat berada pada posisi teratas. Namun, tidak terdapat perbedaan pendapatan yang sangat signifikan antar negara, yang menunjukkan bahwa aktivitas transaksi tersebar cukup seimbang secara geografis.

# **7. Segmentasi Pelanggan (K-Means Clustering)**

```python
df_cluster = pd.read_csv("cluster.csv")

plt.figure(figsize=(6,4))
plt.scatter(
    df_cluster["Avg_Frequency"],
    df_cluster["Avg_Total_Spend"]
)

for i, row in df_cluster.iterrows():
    plt.text(
        row["Avg_Frequency"],
        row["Avg_Total_Spend"],
        f"Cluster {int(row['prediction'])}"
    )

plt.title("Segmentasi Pelanggan Menggunakan K-Means")
plt.xlabel("Rata-rata Frekuensi Transaksi")
plt.ylabel("Rata-rata Total Pembelanjaan")
plt.grid(True)
plt.show()
```

Visualisasi ini menampilkan hasil segmentasi pelanggan menggunakan algoritma K-Means. Setiap titik merepresentasikan satu cluster pelanggan berdasarkan rata-rata frekuensi transaksi dan total pembelanjaan. Cluster dengan nilai rata-rata tertinggi menunjukkan pelanggan dengan loyalitas dan nilai ekonomi yang tinggi, sehingga dapat menjadi target utama dalam strategi pemasaran.
